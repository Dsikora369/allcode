##Завантаження даних##

```python
drive.mount('/gdrive')
%cd /gdrive
path = '/gdrive/My Drive/Colab Notebooks/Datafiles/Hepatitis/hepatitis.data'
names = ['Class', 'AGE', 'SEX', 'STEROID', 'ANTIVIRALS', 'FATIGUE', 'MALAISE', 'ANOREXIA',
         'LIVER BIG', 'LIVER FIRM', 'SPLEEN PALPABLE', 'SPIDERS', 'ASCITES', 'VARICES',
         'BILIRUBIN', 'ALK PHOSPHATE', 'SGOT', 'ALBUMIN', 'PROTIME', 'HISTOLOGY']

data = pd.read_csv(path, names=names, header=None)
data
```

##Обробка даних##

Спочатку перевіримо на наявність відсутніх значень.

```python
data.isna().sum()
```

```python
for name in names:
  print(data[name].unique())
```

Відсутніх значень немає, але видно що у таблиці є "?". Для розуміння того що робити далі треба перевірити скільки їх взагалі і де вони є.

```python
for name in names:
  print(name)
  print(data[data[name]=='?'])
```

Як ми бачимо, є три типи поганих даних. Перший, це рядки в яких просто відсутня більша кількість інформації, а саме 6,7, 14, 31, 41, 56, 72, 83, 118, 147. Колонка "PROTIME" по якій половина даних відсутня, та колонки в яких відсутня не критична кількість даних і ці дані можна замінити модою. Почнемо з видалення поганих рядків.

```python
data1 = data.drop([6 , 7, 14, 31, 41, 56, 72, 83, 118, 147])
```

```python
for name in names:
  print(name)
  print(data1[data1[name]=='?'])
```

Стало набагато краще, тепер видалимо колонку "PROTIME".

```python
del data1["PROTIME"]
```

```python
names.pop(18)
```

```python
for name in names:
  print(name)
  print(data1[data1[name]=='?'])
```

Тепер замінимо відсутні дискретні значення модою, а неперервні сердніми. 

```python
data1['STEROID'][data1['STEROID']=='?'] = data1['STEROID'].mode()[0]
data1['LIVER BIG'][data1['LIVER BIG']=='?'] = data1['LIVER BIG'].mode()[0]
data1['LIVER FIRM'][data1['LIVER FIRM']=='?'] = data1['LIVER FIRM'].mode()[0]
data1
```

```python
for name in names:
  print(name)
  print(data1[data1[name]=='?'])
```

Тепер замінимо середніми.

```python
for column in data1[['BILIRUBIN', 'ALK PHOSPHATE', 'ALBUMIN']]:
  df1 = data1[column][data1[column] != '?']
  df2 = df1.apply(pd.to_numeric)
  data1[column].replace('?', df2.mean(), inplace = True)
```

```python
for name in names:
  print(name)
  print(data1[data1[name]=='?'])
data1
```

Тепер трансформуємо усе у неперервні величини.

```python
data1 = data1.apply(pd.to_numeric)
for name in names:
  print(data1[name].unique())
```

##Розділення даних##

```python

from sklearn.model_selection import train_test_split
```

```python
X = data1.drop(['Class'], axis=1)
y = data1['Class']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=20)

X_train
```

##Вибираємо модель##

```python
from sklearn.metrics import accuracy_score
models =[]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print('LogisticRegression: ', accuracy_score(y_test,y_pred)*100)
models.append(['LogisticRegression: ',accuracy_score(y_test,y_pred)*100])
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(3)
model.fit(X_train, y_train)
train_predictions = model.predict(X_test)
y_pred=model.predict(X_test)
print('KNN3: ', accuracy_score(y_test,y_pred)*100)
models.append(['KNN3 ',accuracy_score(y_test,y_pred)*100])
```

```python
model = KNeighborsClassifier(4)
model.fit(X_train, y_train)
train_predictions = model.predict(X_test)
y_pred=model.predict(X_test)
print('KNN4: ', accuracy_score(y_test,y_pred)*100)
models.append(['KNN4 ',accuracy_score(y_test,y_pred)*100])
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print('XGBClassifier: ', accuracy_score(y_test,y_pred)*100)
models.append(['XGBClassifier: ',accuracy_score(y_test,y_pred)*100])
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=50, random_state=42)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print('RandomForestClassifier: ', accuracy_score(y_test,y_pred)*100)
models.append(['RandomForestClassifier: ',accuracy_score(y_test,y_pred)*100])
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print('SVC: ', accuracy_score(y_test,y_pred)*100)
models.append(['SVC: ',accuracy_score(y_test,y_pred)*100])
```
